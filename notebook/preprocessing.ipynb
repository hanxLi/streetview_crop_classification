```python
import os
import geopandas as gpd
import pandas as pd
import shutil
```

### Rename Iowa&Nebraska

```python
path = "/home/hanxli/data/data/all_sv_imgs"
img_names = [f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]
img_names[0:3]
```

```python
for idx, name in enumerate(img_names):
    os.rename(os.path.join(path, name), os.path.join(path, 'IMG_2022'.join(f"_{idx+1:03d}.jpg")))
```

```python
gdf_2022 = gpd.read_file("/home/hanxli/data/data/labeled_imgs_2022.json", driver='GeoJSON')
gdf_2023 = gpd.read_file("/home/hanxli/data/data/labeled_imgs_2023.json", driver='GeoJSON')
```

```python
gdf_2022.head()
```

```python
path = "/home/hanxli/data/data/IowaNebraska2022"
saves = "/home/hanxli/data/data/all_sv_imgs"

for i in range(len(gdf_2022)):
    filename = gdf_2022.loc[i, "Name"]
    new_name = f"IMG_2022_{i+1:03d}.jpg"
    gdf_2022.loc[i, "Name"] = new_name
    gdf_2022.loc[i, "save_path"] = os.path.join(saves, new_name)
    shutil.copy(os.path.join(path, filename), gdf_2022.loc[i, "save_path"])
```

```python
gdf_2022.head()
```

```python
gdf_2022.to_file("/home/hanxli/data/data/labeled_imgs.json", driver="GeoJSON")
```

### Rename GSV and remove duplicates

```python
esa_gsv_img = gpd.read_file("/home/hanxli/data/data/esa_gsv_img_shp/esa_gsv_img.shp")
esa_gsv_img.head()
```

```python
temp_df = esa_gsv_img.drop_duplicates(subset=['geometry']).reset_index(drop=True)
temp_df
```

```python
path = "/home/hanxli/data/data/GSV_images"
saves = "/home/hanxli/data/data/all_sv_imgs"

for i in range(len(temp_df)):
    filename = temp_df.loc[i, "img_name"]
    new_name = f"IMG_2021_{i+1:04d}.jpg"
    temp_df.loc[i, "img_name"] = new_name
    temp_df.loc[i, "save_path"] = os.path.join(saves, new_name)
    if not os.path.isfile(os.path.join(saves, new_name)):
        shutil.copy(os.path.join(path, filename), temp_df.loc[i, "save_path"])
```

```python
temp_df
```

```python
temp_df.to_file("/home/hanxli/data/data/esa_gsv_labeled_imgs.json", driver="GeoJSON")
```
